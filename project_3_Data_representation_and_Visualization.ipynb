**<h2>Animated and Interactive plots**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

**Sunburst plot to show heirarchy on Titanic dataset**

```python
# used when there are more catorgorical values
df=sns.load_dataset("titanic")
df.head()
```

```python
fig=px.sunburst(df,path=['sex','class','who','alive','alone'],values='survived',color='sex')
fig.show()
```

**draw interactive plots from data plotly**

```python
# load data from plotly
plt_Data=px.data.gapminder()
plt_Data.head()
```

```python
# draw the sunburst chart
schart=px.sunburst(plt_Data,path=['continent','country','year'],values='pop', color='continent')
schart.show()
```

**animated plot**

```python
fig3=px.scatter(plt_Data,x='gdpPercap',y='lifeExp',size='gdpPercap',color='continent',animation_frame='year',animation_group='country',log_x=True)
fig3.show()
```
