**Import all the neccessary Library**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

**Import the Datset**

```python
data=pd.read_csv('/content/Expanded_data_with_more_features.csv')
data.head()
```

**Describing about the data**

**Information About the Data**

```python
data.isnull().sum()
```

**Drop Unnamed Column**

```python
data=data.drop('Unnamed: 0',axis=1)
data.head()
```

```python
print(data.head())
```

**Change Weekly Study hours columns**

```python
data['WklyStudyHours']=data['WklyStudyHours'].str.replace('05-Oct','5-10')
data.head()
# Hence columns can be changed sucessfully
```

**Task.1.<h2>Gender Distribution</h2>**

```python
plt.figure(figsize=(5,4))
x=sns.countplot(data=data,x='Gender')
x.bar_label(x.containers[0])
plt.title('Gender Distribution')
plt.show()
print(data['Gender'].value_counts())
```

**From this Gender Distrubution we anlayses that females count is more than Males count**

**Task.2.<h2>Impact of Parent's Eduction on Students</h2>**

```python
impact=data.groupby('ParentEduc')[['MathScore','ReadingScore','WritingScore']].mean()
impact
```

```python
# Represent it into graphical representation
sns.heatmap(impact)
```
